```python
# Start Angle
# Default start angle is at x-axis
# Start with the first wedge at 90 degrees
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle=90)
plt.title("Simple Pie Chart (first wedge starting at 90 degreess)")
plt.show()

# You can see the first wedge (Apples) starts at 90 degrees and moves counterclockwise
```

```python
# Explode
# Pull the "Apples" wedge 0.2 from the center of the pie
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode=myexplode)
plt.title("Simple Pie Chart (Apples wedge exploded 0.2 from center)")
plt.show()
```

```python
# Sales data from sales.csv
# amount sold for each products
# bar chart

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales.csv')

df = df.groupby('product', as_index=False)['count'].sum()

bars = plt.bar(df['product'], df['count'], width = 5, color ='CornflowerBlue')

plt.bar_label(bars, label_type='edge')

plt.title('Amount sold for each product')  
plt.xlabel('Amount Sold')
plt.ylabel('Products')  
plt.show()

"""
groupby() here groups thedata by unique values in the product columns
as_index is used so product column is not used as index
"""
```

```python
# Sales data from sales.csv
# amount sold for each products
# pie chart
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('sales.csv')
df = df.groupby('product', as_index=False)['count'].sum()

mycolors = ["CornflowerBlue", "DarkSeaGreen", "PaleVioletRed", "Crimson"]
myexplode = [0.15, 0, 0, 0]

plt.pie(df['count'], labels=df['product'], colors=mycolors, autopct='%1.1f%%', explode=myexplode, shadow=True)
# make sure fist argument to pie() is numeric data

plt.title('Percentage Sold Each Product')
plt.legend(loc='upper left', title='Products')
plt.show()
```

```python
# Line graoph
import pandas as pd

df = pd.read_csv('sales.csv')

df['time'] = pd.to_datetime(df['time'], errors='coerce')

df['Month'] = df['time'].dt.month

# aggregate counts by month / add the counts for each month using sum()
montly_data = df.groupby('Month', as_index=False)['count'].sum()

plt.plot(montly_data['Month'], montly_data['count'], marker='o')
plt.title('Products Sold by Month')
plt.xlabel('Months')
plt.ylabel('Amount Sold')
plt.xticks(ticks=range(13), labels=[' ', 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October',
                                    'November', 'December'], rotation=75)
plt.grid(True)
plt.show()
```
