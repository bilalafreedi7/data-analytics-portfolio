```python
# 1) Training
# a) b)
import pandas as pd
df = pd.read_csv('diabetes.csv')

df.head()
```

```python
# 1) Training
# c) Create array of dependent (Outcome) and independent variable
x = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
# 1) Training
# d) Create training data (70% training and 30% testing)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

```

```python
# 2) Fit the model and testing
# a) Fit the model with training data (Decision Tree with max_depth=3)
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

model = tree.DecisionTreeClassifier(max_depth=3)
model = model.fit(x_train, y_train)
tree.plot_tree(model)
```

```python
# 2) Fit the model and testing
# b) Test the model by predicting testing data

y_pred = model.predict(x_test)

comparison = pd.DataFrame({'Real': y_test, 'Predicted': y_pred})
print(comparison.head())
```

```python
# 3) Evaluate the model
# a) Compare result of prediction from testing data and real data of testing data
# b) Display evaluating score of accuracy, precision, and recall
from sklearn.metrics import *

print('accuracy: ', accuracy_score(y_test, y_pred))
print('precision: ', precision_score(y_test, y_pred))
print('recall: ', recall_score(y_test, y_pred))
```
