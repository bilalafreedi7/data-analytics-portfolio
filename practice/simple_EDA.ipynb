```python
# Rename some of our columns with bad names using rename()
df = df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'})

# Utilize a dictionary to do this
```

```python
# Identify where missing values exist in the dataframe and how often they occur
# Use isna() to identify missing/null values
df.isna()

# When using isna() if 'True' is in the cell then that cell is null
# This is hard to look at when identifying null values
```

```python
# Checking an example duplicate
# Test and look for specific coaster names and what differences their rows have
df.query('Coaster_Name == "Crystal Beach Cyclone"')

"""
Explanation of 'df.query('Coaster_Name == "Crystal Beach Cyclone"')'

This line filters the dataframe (df) and returns only the rows where the value in the
'Coaster_Name' column is equal to 'Crystal Beach Cyclone'.
Could also use df[df['Coaster_Name'] == "Crystal Beach Cyclone"]

Definition of query():
This function in pandas is used to filter rows in a DatFrame by evaluating a query string.
This method provides a cleaner and more readable way to filter data compared to using
boolean indexing or complex conditional statements.
"""

# Here, the 'Year_Introduced' is different
```

```python
df.columns
```

```python
# Remove any duplicates with certain number of columns that are the same
# Run this on a subset of columns
# If coaster name, location, and opening date are duplicated, identify them
# Use sume to see how many
# Use '~' to take inverse and select just columns that are NOT the duplicate, the first versions
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])] \
    .reset_index(drop=True).copy()
    
# reset_index() will reset the index after duplicated rows are dropped

# We can see 97 rows are duplicated when sum() is added
```

```python
# STEP 3: Feature Understanding
# (Univariate analysis)
# Plotting Feature Distributions
    # Histogram
    # KDE (Kernel Density Estimation)
    # Boxplot
    
"""
Univariate Analysis Explanation in EDA

This is the simplest form of data analysis, where only one variable is analyzed at a
time. The goal is to understand and summarize the basic characteristics of the variable,
such as its distribution, central tendency, disperson, and spread.    
"""
df['Year_Introduced'].value_counts()

# value_counts() looks for any duplicates and counts how many unique values occur in dpecified column
# Automatically order from most to least occuring
```

```python
# Turn previous code block results into a plot to see top years of coasters introduced
# Only use the top 10 mots common years with most coasters introduced
# Use a bar plot (kind='bar')
import matplotlib.pyplot as plt
df['Year_Introduced'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Years Coasters Introduced')
plt.xlabel('Year Introduced')
plt.ylabel('Count')
```

```python
# Something lese to do when analyzing one column of data is to get idea of distribution of the column
# Speed_mph as example
# Create plot to show distribution, use a histogram (kind='hist')
# Use 20 bins (bins=20) to show greater detail of distribution
speed_dist = df['Speed_mph'].plot(kind='hist', bins=20)
plt.title('Coaster Speed (mph)')
plt.xlabel('Speed (mph)')
```

```python
# Create histogram for other features to see distribution
# Distribution of 'Height_ft' column
height_dist = df['Height_ft'].plot(kind='hist', bins=25)
plt.title('Coaster Height (ft)')
plt.xlabel('Height (ft)')

# We can see most coasters have heights less than 150
# There are outliers, coasters with heights around 320 and 370
```

```python
# Check distribution for 'Year_Introduced' column
year_dist = df['Year_Introduced'].plot(kind='hist', bins=20)
plt.title('Year Introduced')
plt.xlabel('Year')

# Most coasters introduced after 1990
# Huge spike in coasters introduced late 1990s into 2000s
```

```python
# We can also use a density plot (kde) to show distribution in a column/feature
speed_dist2 = df['Speed_mph'].plot(kind='kde')
plt.title('Coaster Speed (mph)')
plt.xlabel('Speed (mph)')
```

```python
# STEP 4: Feature Relationships
    # Scatterplot
    # Heatmap Correlation
    # Pairplot
    # Groupby comparisons
    
# Scatterplot
df.plot(kind='scatter', x='Speed_mph', y='Height_ft')
plt.title('Coaster Speed vs. Height')
plt.show()

# This shows a scatter plot that shows the relationship between Speed_mph and Height_mph
```

```python
# Use seaborn for most advanced plots
# Use Year_Introduced as hue
sns.scatterplot(x='Speed_mph', y='Height_ft', hue='Year_Introduced', data=df)
plt.title('Coaster Speed vs. Height')
plt.show()

# Using hue parameter shows each data point and its year, and adds a legend
```

```python
# Pairplot
# Comparing more than two features
sns.pairplot(df, vars=['Year_Introduced', 'Speed_mph', 'Height_ft', 
                       'Inversions', 'Gforce'], hue='Type_Main')
plt.show()

# Shows distribution and relationship in a matrix style
# Uses 'Type_Main' (Type of material) as hue to group coasters by material stype
```

```python
# Show correlation between different features

df_corr = df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
print(df_corr)
"""
Explanation of code above:

dropna() removes rows or columns containing missing values from a DataFrame
    - Paramters: axis, how (used with 'any' or 'all'), subset, inplace.
    
corr() calculates pairwise correlation between numeric features of a DataFrame
    - Parameters: method (pearson, kendall, spearman), min_periods (minimum # of observations)
    - 1: Perfect positive correlation
    - -1: Perfect negative correlation
    - 0: No correlation
    
'Speed_mph' and 'Height_ft' have acorrelation of .74, indicating a positive linear relationship

"""
```

```python
# Heatmaps
# Use seaborn heatmaps to show correlations
sns.heatmap(df_corr, annot=True)
plt.title('Correlations between data')
plt.show()

# annot=True shows raw values of the correlations
```

```python
# STEP 5: Ask a Question about the data
    # Try to answer a question you have about the data using a plot or statistic.

# What are the locations with the fastest roller coasters (minimum of 10)?
avg_speed_by_location = df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']) \
    .query('count >= 10').sort_values('mean')['mean'].plot(kind='barh', figsize=(12, 5))

plt.title('Average Coaster Speed by Location')
plt.xlabel('Average Coaster Speed')
plt.show()

# Shows parks with highest average coaster speeds
```
