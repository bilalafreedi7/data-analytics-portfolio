```python
# Rotten Tomato Movie Ratings
# Import
import pandas as pd

df = pd.read_csv("Rotten_Tomatoes_Movies.csv")

df.shape
```

```python
df.columns
```

```python
# check what columns have null values
df.isnull().sum()
```

```python
# check datatypes
df.dtypes
```

```python
# convert date columns to datetime
df['in_theaters_date'] = pd.to_datetime(df['in_theaters_date'], errors='coerce')
df['on_streaming_date'] = pd.to_datetime(df['on_streaming_date'], errors='coerce')

```

```python
df.head()
```

```python
df['theater_year'] = df['in_theaters_date'].dt.year
df['streaming_year'] = df['on_streaming_date'].dt.year

df
```

```python
top_genres = (df.groupby('genre')['audience_count'].sum().sort_values(ascending=False).head(10))
print(top_genres)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_genres = df.groupby('genre')['audience_count'].sum().n(10).sort_values(ascending=False)
top_genres_df = top_genres.reset_index()
sns.countplot(y='genre', data=df[df['genre'].isin(top_genres_df['genre'])], order=top_genres_df['genre'])
plt.title('Top 10 Genres by Audience Count')
plt.xlabel('Audience Count')
plt.ylabel('Genre')
plt.show()
```

```python
df['audience_rating'].hist(bins=20, figsize=(8, 6))
plt.title('Distribution of Audience Rating')
plt.xlabel('')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_columns = ['tomatometer_rating', 'runtime_in_minutes', 'audience_rating', 'tomatometer_count', 'audience_count']

correlation_data = df[numerical_columns]

correlation_matrix = correlation_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", linewidths=2)
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```
