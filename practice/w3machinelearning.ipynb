```python
# Mean
import numpy as np

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = np.mean(speed)

print(x)
```

```python
# Median
import numpy as np

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = np.median(speed)

print(x)
```

```python
# Mode
from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = stats.mode(speed)

print(x)
```

```python
# Standard Devaition
import numpy as np

speed = [86,87,88,86,87,85,86]

x = np.std(speed)

print(x)
```

```python
# Variance
import numpy as np

speed = [32,111,138,28,59,77,97]

x = np.var(speed)

print(x)

"""
A varianceof 1432.24 indicates that the speed measuremens are quite
spread out from the mean. The higher the variance, the more spread out
the data points are.

If there was a another set of speed measurements with a lower variance,
it owuld mean those speeds were more consistent/clustered around their
mean.
"""
```

```python
# Percentile
import numpy as np

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = np.percentile(ages, 75)

print(x)
```

```python
# Data Distribution
# create an array containing 250 random floats between 0 and 5, two decimal places
import numpy as np

x = np.round(np.random.uniform(0.0, 5.0, 250), 2)

print(x)

"""
About np.random.uniform()

It is a function that genrates random numbers from a uniform distribution.
In the above code, there is equal probability of every number between 
0 and 5 (equal chance of being selected)

Parameters:
- low: minimum possible value
- high: maximum possible value
- size: how many numbers to generate

Contrast with other distributions:
- Normal distribution: values cluster around the mean
- Uniform distribution: no clustering - completely flat probability

Use when you want a fair randomness across a range.
"""
```

```python
# Histogram
# draw a histogram with uniform distribution
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0.0, 5.0, 250)

plt.hist(x, 5) # 5 bars
plt.show()

"""
Explanation:

The first bar represents how many values in the array are between 0 and 1.
The second bar represents how many values are between 1 and 2.
etc.
"""
```

```python
# Histogram
# draw a histogram with normal distribution
import numpy as np
import matplotlib.pyplot as plt

x = np.round(np.random.normal(100, 10, 250), 2)
# generate 250 values, concentrating aorund 100 with standard deviation as 10

plt.hist(x)
plt.show()
```

```python
# Big Data Distribution (uniform)
# create an array with 100000 random numbers from 0 to 5
# display them using a histogram with 100 bars
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0.0, 5.0, 100000)

plt.hist(x, 100) # 100 bars
plt.show()
```

```python
# Big Data Distribution (normal)
# create an array with 100000 random numbers
# from 0 to 5 clustered around 2.5
# with standard deviation of 1.5
# display them using a histogram with 100 bars
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(2.5, 1.5, 100000)

plt.hist(x, 100) # 100 bars
plt.show()
```

```python
# Scatter Plot
# Use the scatter() method to draw a customized scatter plot diagram
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y, 
            s=25, 
            c='royalblue', 
            edgecolors='black', 
            linewidths=.7, 
            marker='d')

plt.grid(True, ls='--')
plt.title('Customized Scatter Plot',
          fontsize = 18, 
          pad=20) # pad is distance from plot

plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()

"""
Explanation: 

Let's assume the x-axis represents age and the y-axis represents speeds.
What we can read from this scatter plot is that the two fastest cars were
both 2 years old, and the slowest car was 12 years old. 

It seems that the newer the car, the faster it drives, but that could be
a coincidence, after all, we only have 13 cars in this data.
"""
```

```python
# Random Data Distributions
# A scatter plot with 1000 dots
# create two arrays with 1000 random numbers with normal data distribution
# first array has mean as 5, standard deviation of 1
# second array has mean at 10, standard deviation of 2
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)

plt.scatter(x, y,
         s=25,
         c='royalblue',
         marker='d')

plt.grid(True, ls='--')

plt.title('Scatter Plot with 1000 Dots',
          fontsize=18,
          pad=15)
plt.xlabel('X Vlaues', fontsize=14)
plt.ylabel('Y Values', fontsize=14)

plt.tight_layout()
plt.show()

"""
Explanation:
We can see that the dots are connected srounf the value 5 on the x-axis,
and 10 on the y-axis.

We can also see that the spread is wider on the y-axis than on the x-axis
since the standard deviation of the Y values is higher than of the X values.
"""
```
