```python
# Linear Regression
"""
In the example below, the x-axis represents age, and the y-axis represents
speed. Regstered are the age and speed of 13 cars as they were passing a
tollbooth. See if the data collected could be used in a linear regression.
"""
# Start by drawing a scatter plot
import matplotlib.pyplot as plt
from scipy import stats # import scipy to draw line of linear regression
 
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept
    
mymodel = list(map(myfunc, x))

plt.scatter(x, y,
            s=30,
            c='royalblue',
            marker='d')
plt.grid(True, ls='--')
plt.title('Age and Speed of Cars', fontsize=15)
plt.xlabel('Age')
plt.ylabel('Speed')
plt.plot(x, mymodel)
plt.show()

print('r: ', r)
print('p: ', p)
print('std_err: ', std_err)
print('A list of predicted y-values based on the regression line for each x-value: \n', mymodel)

"""
Explanation of code:

Null Hypothesis:
'There is no linear relationship between x and y, or slope is 0'

'slope, intercept, r, p, std_err = stats.linregress(x, y)'
- This line calculates linear regression statistics using stats.linregress():

    - slope: the slope of the regression line (how much y changes for each
             unit change in x)
             
    - intercept: the y-intercept of the regression line (value of y when x=0)
    
    - r: the correlation coefficient (measures strength and direction of
         linear relationship, -1 to 1).
         r ~ 1 strong positive correlation
         r ~ 0.5 modersate positive correlation
         r ~ 0 no linear relationship
         same for -1 and -0.5 except its negative
         In this case it is -0.76, showing that there is relationship, not
         perfect, but it indicates that we could use linear regression in 
         future predictions.
         
    - p: the p-value for a hypothesis test whose null hypothesis is that
         the slope is zero. A p-value is basically the likelihood of finding
         a mean difference by chance if indeed there is no difference in the 
         population.  
         In this case the p-value is well below 0.5 meaning there is strong
         evidence to reject the null hypothesis, meaning that the relationship
         between x and y is statistically significant.
          
    - std_err: standard error of the estimated slope
    
'def myfunc(x):
    return slope * x + intercept'
- a function that calculates the predicted y-value for any given x using
  the regression line equation (y = slope * x + intercept)
  
'mymodel = list(map(myfunc, x))'
- map(myfunc, x): applies the myfunc function to each element in the x list

- list() converts the map object to a list

- mymodel: a list of predicted y-values based on the regression line for
  each x-value.
"""
```

```python
# CONTINUATION OF LAST CODE BLOCK
# Predict Future Values
"""
Now we can use the information we have gathered to predict future values.
Example: Let us try to predict the speed of a 10 year old car.
To do this, we need the same myfunc() function from code block above.
"""
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

# predict speed of 10 year old car
speed = myfunc(10)

print(round(speed, 2), 'mph')

# The predicted speed comes out to be 85.6
# This would be (10, 85.6) on the regression line
```

```python
# Bad Fit
# Example where linear regression wouldn't be the best method to predict future values
# The values for the x-axis and y-axis should result in a very bad fit
import matplotlib.pyplot as plt
from scipy import stats

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y,
            s=25,
            c='royalblue',
            marker='d')
plt.grid(True, ls='--')
plt.title('Bad Fit Regression Line Example', fontsize=15, pad=10)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.tight_layout()
plt.plot(x, mymodel)
plt.show()

print('r: ', r)
print('p-value: ', p)

"""
Low r value of 0.013 indicates a very bad relationship, and tells use that
this dats set is not suitable for linear regression.

The p-value of 0.95 is high meaning that we can accept the null hypothesis
that there is no relationship between x and y. There is no linear relationship.
"""
```

```python
# Polynomial Regression
"""
In the example below, registered are 18 cars as they were passing a certain
tollbooth. Registered are the car's speed, and the time of day (hour) the
passing occurred.
The x-axis represents the hours of the day and the y-axis represents the
speed.
"""
# Start by drawing a scatter plot
import matplotlib.pyplot as plt
import numpy as np    # to draw line of polynomial regression

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = np.poly1d(np.polyfit(x, y, 3)) # polyfit 

myline = np.linspace(1, 22, 100)

plt.scatter(x, y,
            s=25,
            c='royalblue',
            marker='d')
plt.grid(True, ls='--')
plt.title('Speed of Cars at Different Hours', fontsize=15, pad=10)
plt.xlabel('Hours of the day')
plt.ylabel('Speed')
plt.tight_layout()
plt.plot(myline, mymodel(myline))  # plots the fitted polynomial, myline are x points
plt.show()

"""
Explanation:

'mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))'
- 'np.polyfit(x, t, deg)
    - fits a polynomial of degree 'deg' to the data (x, y)
    - returns: coefficient of the polynomial (highest dgree first)
    - degree (3 in this case) means a cubic polynomial
    
- 'np.poly1d(coefficients)
    - creates a polynomial function from coefficients
    - mymodel is now a callable function that predicts y for any x.

'myline = np.linspace(1, 22, 100)'
- np.linespace(start, stop, num)
    - generates 100 evenly spaced points between 1 and 22
    - used to plot a smooth curve (instead of jagged lines between
      original x points)
"""
```

```python
# CONTINUATION OF PREVIOUS CODE BLOCK
# Polynomial Rrgression
# R-squared --> how well the model explains variance in y
# How to evaluate the fit?
# Calculating R-Squared
from sklearn.metrics import r2_score

r2 = r2_score(y, mymodel(x))

print("R-squared:", r2)

"""
r2_score(y_true, y_pred)
Explanation: 

'mymodel(x)
What it does:
- uses the fitted polynomial (mymodel) to predict y values for all x inputs
- Example: if mymodel = 2x + 3x + 1, then mymodel([1, 2]) returns [6, 15]

'r2_score(y, mymodel(x))'
What happens:
1. computes predicted y values: y_pred = mymodel(x)
2. compares y_pred with true y values using the r-squared formula
3. returns a single number

The result, or r-squared comes out to be 0.94, which shows that there is a very good
relationship, and we can use polynomial regression in future predictions.

"""
```

```python
# Scale
# Predict CO2 Values
# Predict the CO2 emission from a 1.3 liter car that weighs 2300 kilograms
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

df = pd.read_csv("Car_CO2_Data.csv")

X = df[['Weight', 'Volume']]
y = df[['CO2']]

scaledX = scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaledX, y)

scaled = scale.transform([[2300, 1.3]])

predictedCO2 = regr.predict([scaled[0]])
print(predictedCO2)

"""
Explanation of code:

This code predicts the co2 emmissions of a car based on its weight (2300kg) and engine volume
(1.3 liters) using linear regression with feature scaling. 

First the code read the data, then puts Weight and Volume (independent variable) in a datafram
called X. Then puts CO2 (dependent variable) in a dataframe called y.

'scale  = StandardScaler()'
    - creates a StandardScaler object named scale
    - Why? To standardize features (Weight and Volume) so they contribute equally to the model

'scaledX = scale.fit_transform(X)'
    - fit_transform(X) does two things:
        fit(): Calculates the mean and standard deviation of Weight and Volume
        transform(): Applies scaling using z = (x - u) / s
    - Why scale? Prevents Weight (e.g., 1000-3000 kg) from dominating Volume (e.g, 1.0-2.0 L)
      in the model.
      
'regr = linear_model.LinearRegression()
 regr.fit(ScaledX, y)'
    - LinearRegression() initializes a linear regression model
    - fit(scaledX, y) trains the model to find the best-fit line using:
        scaledX (scaled features)
        y (actual co2 values)
    - What happens? the model learns:
        Coefficients (regr.coef_): How much Weight and Volume affect CO2
        Intercept (regr.intercept_): Baseline co2 when both features are zero
        
'scaled = scale.transform([[2300, 1.3]])'
    - transform() svales the enw cra data ([2300 kg, 1.3L]) using the same mean/std from training.
    - Why? Ensures the new input is consistent with the training data.
    
'predictedCO2 = regr.predict([scaled[0]])'
    - predict() estimates CO2 using the trained model.
    Input: Scaled values of [2300, 1.3]
    Output: Predicted CO2 [[107.2]]
"""
```
